🔗 What is Correlation?

Correlation measures the strength and direction of a linear relationship between two numerical variables.

📌 Example:
As temperature increases, humidity decreases → Negative correlation

As rainfall increases, soil moisture increases → Positive correlation

📐 How is It Calculated?

Using Pearson correlation coefficient by default:

```python
# check the correlation of the features
numeric_columns = crop.select_dtypes(include=[np.number])
corr = numeric_columns.corr()
corr
```

🧮 What is a Correlation Matrix?

A correlation matrix is a table showing correlation values between all pairs of numerical variables.

🧠 Example Interpretation:

In the correlation matrix:

If N and P have a correlation of 0.8, they strongly increase together.

If temperature and humidity have a correlation of -0.7, one increases while the other decreases.

If a correlation is near 0, they don’t affect each other linearly.

```python
# Visualize the correlation matrix
import seaborn as sns
sns.heatmap(corr,annot=True,cbar=True,cmap='coolwarm')
plt.title("Heatmap of Features")
plt.show()
```

## Encoding

🔤 What is Encoding?

Encoding is the process of converting categorical (text) data into numerical format so that it can be used by machine learning models.

📌 Types of Encoding:

1. Label Encoding: Converts each category into a unique number.

2. One-Hot Encoding: Creates a new column for each category, marking presence with 1 or 0.


🔁 What is Decoding?
✅ Definition:
Decoding is the reverse of encoding — converting the numerical values back to their original categories.

```python
crop_dict = {
'rice':1,
'maize':2,
'jute':3,
'cotton':4,
'coconut':5,
'papaya':6,
'orange':7,
'apple':8,
'muskmelon':9,
'watermelon':10,
'grapes':11,
'mango':12,
'banana':13,
'pomegranate':14,
'lentil':15,
'blackgram':16,
'mungbean':17,
'mothbeans':18,
'pigeonpeas':19,
'kidneybeans':20,
'chickpea':21,
'coffee': 22,
}
```

```python
crop
```

```python
crop['crop_no'] = crop['label'].map(crop_dict)
```

```python
crop.head()
```

```python
crop.drop('label',axis=1,inplace=True)
crop.tail()
```

## Train Test Spilt

```python
x = crop.drop('crop_no',axis=1)
y = crop['crop_no']

print(f"The shape of x is : {x.shape}")
print(f"The shape of y is : {y.shape}")
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=42)
```

```python
print(f"Shape of x_train: {x_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of x_test: {x_test.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

Training Models

```python
# Let's Train a Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# intialize the DecisionTree Classifier
DT = DecisionTreeClassifier()
# Train the model
DT.fit(x_train,y_train)
```

```python
# evaluate the model on test dataset
y_pred = DT.predict(x_test)
print(f"Accuracy score of Decision Tree Classifier is: {accuracy_score(y_test,y_pred)}")
```

```python
# evaluate the model on train dataset
y_pred_train = DT.predict(x_train)
print(f"Accuracy score of Decision Tree Classifier is: {accuracy_score(y_train,y_pred_train)}")
```

Predictive System

```python
def crop_recommend(N,P,K,temperature,humidity,ph,rainfall):
    features = np.array([[N,P,K,temperature,humidity,ph,rainfall]])
    transformed_features = sc.transform(features)
    prediction = DT.predict(transformed_features).reshape(1,-1)
    crop_dict = {1: "Rice", 2: "Maize", 3: "Jute", 4: "Cotton", 5: "Coconut", 
                 6: "Papaya", 7: "Orange", 8: "Apple", 9: "Muskmelon", 10: "Watermelon", 
                 11: "Grapes", 12: "Mango", 13: "Banana", 14: "Pomegranate", 15: "Lentil", 
                 16: "Blackgram", 17: "Mungbean", 18: "Mothbeans",19: "Pigeonpeas", 
                 20: "Kidneybeans", 21: "Chickpea", 22: "Coffee"}
    crop = [crop_dict[i] for i in prediction[0]]

    
    return f"{crop} is a best crop to be cultivated " 
```

```python
# Predicting the crop for the given environment
N = 120
P = 120
K = 120
temperature = 10
humidity = 55
ph = 6.1
rainfall = 150

crop_recommend(N,P,K,temperature,humidity,ph,rainfall)
```
